```python
import cv2
import numpy as np

cap = cv2.VideoCapture(r"C:\Users\zbook 17 g3\Downloads\video.mp4")
func = cv2.createBackgroundSubtractorMOG2()
cc = []
carros=0
while True:
    ret, frame = cap.read()

    if not ret:
        break

    grey = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(grey, (3, 3), 5)
    
    mask = func.apply(blur)
    kernel = np.ones((2, 2), np.uint8)
    mask=cv2.erode(mask,kernel,iterations=5) 
    contours, hierarchy  =  cv2.findContours(mask, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

    cv2.line(frame, (25, 550), (1200, 550), (176, 130, 39), 2)

    for contour in contours:
        
        x, y, w, h = cv2.boundingRect(contour)
        
        condition = (w >= 80) and (h >= 80)
        

        if not condition:
            continue

        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)
        center = (x + w // 2, y + h // 2)
        cc.append(center)
        cv2.circle(frame, center, 4, (0, 0, 255), -1)

    for (x, y) in cc:
        if (y < (550 + 6)) and (y > (550 - 6)):
            carros += 1
            cv2.line(frame, (25, 550), (1200, 550), (0, 0, 255), 3)
            cc.remove((x, y))
    cv2.putText(frame, "VEHICLE COUNT : "+str(carros), (320, 70),cv2.FONT_HERSHEY_COMPLEX, 2, (0, 0, 255),4)

    cv2.imshow("frame",frame)
    cv2.imshow("mask", mask)

    if cv2.waitKey(1) == 27:
        break

cv2.destroyAllWindows()
cap.release()
    
    
    
    
    
    
    


```
